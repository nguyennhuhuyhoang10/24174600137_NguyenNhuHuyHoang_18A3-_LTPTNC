Sử dụng Getter,Setter,deleter


```python
class Dog:
    def __init__(self, name, size,age,color):
        self._name=name
        self._size=size
        self.__age=age
        self.__color=color
    @property
    def age(self):
        print("Getter for age called")
        if not hasattr(self, "_Dog__age"):
            raise AttributeError("Age has been deleted or not set")
        return self.__age
    @age.setter
    def age(self, value):
        print("Setter for age called")
        if value<0:
            raise ValueError("Age cannot be negative")
        self.__age=value
    @age.__deleter
    def age(self):
        print("Deleter for age called")
        if not hasattr(self, "_Dog__age"):
            raise AttributeError("Age has already been deleted or not set ")
        del self.__age
```

VD1.3.2.2.Đa kế thừa

```python
class Zebra:
    def __init__(self):
        print("Zebra")
    def Display(self):
        print("I'm a zebra")
class Donkey:
    def __init__(self):
        print("Donkey")
    def Display(self):
        print("I'm a donkey")
class Zonkey(Zebra, Donkey):
    def __init__(self):
        Zebra.__init__(self)
        Donkey.__init__(self)
        print("Zonkey")
obj=Zonkey()
obj.Display()
```

```python
class LopCha:
    def __init__(self, ten_cha):
        self.ten_cha=ten_cha
    def thong_tin(self):
        print("Tôi là lớp cha, tên tôi là", self.ten_cha)
class LopCon(LopCha):
    def __init__(self, ten_cha, ten_con):
        super().__init__(ten_cha)
        self.ten_con=ten_con
    def thong_tin(self):
        print("tôi là lớp con, tên tôi là", self.ten_con)
        super().thong_tin()
class LopChau(LopCon):
    def __init__(self, ten_cha, ten_con, ten_chau):
        # Gọi hàm khởi tạo của LopCha và LopCon để khởi tạo các thuộc tính
        # LớpChau kế thừa LopCon
        super().__init__(ten_cha, ten_con)
        self.ten_chau = ten_chau

    def thong_tin(self):
        print("Tôi là Lớp Cháu, tên của tôi là", self.ten_chau)
        # Gọi phương thức của LopCon
        super().thong_tin()


# Sử dụng các lớp
lop_chau = LopChau("Ông A", "Bác B", "Cháu C")
lop_chau.thong_tin()
```
