Ví dụ: So sánh về lưu trữ bộ nhớ giữ List và Numpy Array

```python
import numpy as np  #import thư viện numpy để tạo và thao tác với mảng NumPy. 
import sys  #import thư viên sys để đo kích thước của đối tượng Python bằng sys.getsizeof().
# Tạo một danh sách Python với 1000 phần tử. 
S = list(range(1000))
# Đo kích thước của mỗi phần tử trong danh sách S 
items_S= sys.getsizeof(S)
print("Kích thước mỗi phẫn tử trong List tính theo bytes: ",items_S)
# Đo kích thước của toàn bộ danh sách S 
S_size = items_S*len(S)
print("Kích thước của toàn bộ List tính theo bytes: ",S_size)
# Tạo một mảng NumPy array với 1000 phần tử 
D= np.arange(1000)
# Đo kích thước của mỗi phần tử trong Numpy array
print("Kích thước mỗi phần tử của mảng Numpy theo bytes: ",D.itemsize)
# Đo kích thước của toàn bộ mảng Numpy array 
array_size=D.itemsize*D.size 
print("Kích thước của toàn bộ mảng Numpy array in bytes: ",array_size) 
print(f"Tỷ lệ (list/array): {S_size / array_size:.2f}")
```

HÀM np.linspace()

```python
import numpy as np 
# Tạo mảng chứa 5 giá trị từ 0 đến 10 (gồm giá trị kết thúc) 
arr_1 = np.linspace(0, 10, 5) 
print(arr_1) 
```

```python
# Tạo mảng chứa 10 giá trị từ 0 đến 1 (không chứa giá trị # kết thúc) 
arr_2 = np.linspace(0, 1, 10, endpoint=False) 
print(arr_2)
```

```python
# Tạo mảng chứa 100 giá trị từ 1 đến 10 và trả về kích thước bước. 
arr_3, step = np.linspace(1, 10, 100, retstep=True) 
print(arr_3) 
print("Step size:", step)
```
