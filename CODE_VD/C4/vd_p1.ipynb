vd4.3.4.2

```python
import pandas as pd 
indA=pd.Index([0, 2, 4, 6, 8]) 
indB=pd.Index([1, 3, 7, 8, 9]) 
print('Index A: ', indA) 
print('Index B: ', indB) 
print('Hợp: ', indA.union(indB)) #hợp (union) 
print('Hiệu: ', indA.difference(indB)) #hiệu 
print('Giao: ', indA.intersection(indB))  #Giao 
print('Hiệu đối xứng: ', indA.symmetric_difference(indB))#hiệu đối xứng 
print('A <= B and A!=B: ', indA.isin(indB))
```

4.4.2.DataSelectionvàIndexingtrongDataFrame

```python
import pandas as pd 
population = pd.Series({'Hà Nội':8330830, 
                        'Quảng Ninh':1358490, 
                        'Nam Định': 1771000, 
                        'Đà Nẵng':1191381, 
                        'TP.Hồ Chí Minh':9411805}) 
area=pd.Series({'Hà Nội':3358.9, 
                 'Quảng Ninh':6177.7, 
                 'Nam Định': 1668.0, 
                 'Đà Nẵng':1284.9, 
                 'TP.Hồ Chí Minh':2061.0}) 
data = pd.DataFrame({'Dân số':population, 'Diện tích':area}) 
data
```

vd4.4.2.1

```python
data['Dân số']
```

 Truy cập trực tiếp vào các Series như là một thuộc tính của đối tượng

```python
#’tạo cột mới ‘popolution’ từ cột ‘Dân số’  
data['population'] = data['Dân số']   
print(data)
```

```python
data.population
```

```python
data['Mật độ']=data['Dân số']/data['Diện tích'] 
data
```

### Hoán chuyển cột và hàng

Ví dụ 4.4.2.3.a

```python
#Sử dụng loc 
print(data) 
#Lấy hàng đầu tiên 
print('\n Dữ liệu của Hà Nội:\n',data.loc['Hà Nội']) 
#Lấy dữ liệu 3 hàng đầu tiên 
print("\n Dữ liệu của Hà Nội, Quảng Ninh và Nam Định:\n",data.loc[:'Nam Định']) 
#Dữ liệu của 2 hàng và 2 cột đầu tiên 
print("\nDữ liệu dân số, diện tích của Hà Nội, Quảng Ninh:\n",data.loc[:'Quảng Ninh',:'Diện tích']) 
#Lấy dữ liệu của tất cả các hàng và các cột cuối cùng 
print("\n Mật độ dân số các Tỉnh/Thành phố:\n",data.loc[:,'Mật độ'])
```
