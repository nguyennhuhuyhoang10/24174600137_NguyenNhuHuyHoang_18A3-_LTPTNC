4.9 Pivot Table & Crosstab trong Pandas

Load cơ sở dữ liệu Titanic vf hiển thi 5 mẫu đầu tiên

```python
import pandas as pd
import seaborn as sns

# Nạp dữ liệu Titanic
titanic_data = sns.load_dataset("titanic")

# Hiển thị một số mẫu dữ liệu
titanic_data.head(10)
```

4.9.1.1 Đặt vấn đề

```python
titanic_data.groupby("sex")["age"].mean()
```

tổng hợp dữ liệu bằng GroupBy (theo TLHT)

```python
gb_age = titanic_data.groupby(["sex", "class"])["age"].mean()
gb_age_unstack = gb_age.unstack()
gb_age_unstack
```
