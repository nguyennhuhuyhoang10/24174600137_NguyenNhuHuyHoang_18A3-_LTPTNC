```python
import pandas as pd

ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
#1
ser1_not_in_ser2 = ser1[~ser1.isin(ser2)]
print("1. Các phần tử của ser1 không nằm trong ser2:")
print(ser1_not_in_ser2)

#2
ser2_not_in_ser1 = ser2[~ser2.isin(ser1)]
ser_symmetric_diff = pd.concat([ser1_not_in_ser2, ser2_not_in_ser1])

print("\n2. Các phần tử chỉ xuất hiện ở ser1 hoặc ser2 (không trùng nhau):")
print(ser_symmetric_diff)
```

```python
import pandas as pd
import numpy as np
ser = pd.Series(np.random.normal(10, 5, 25))

# 1
min = ser.min() 
# 2
p25 = ser.quantile(0.25) 

# 3
median = ser.median() 

# 4
p75 = ser.quantile(0.75)

# 5
max = ser.max() 
print("--- Các thống kê cơ bản ---")
print(f"1. Giá trị tối thiểu (Min): {min:.2f}")
print(f"2. Phân vị thứ 25 (P25): {p25:.2f}")
print(f"3. Trung vị (Median/P50): {median:.2f}")
print(f"4. Phân vị thứ 75 (P75): {p75:.2f}")
print(f"5. Giá trị tối đa (Max): {max:.2f}")
```

```python
import pandas as pd
emails = pd.Series(['I.buying books at amazom.com', 'rameses@egypt.com', 'matt@t.co', 'narenda@modi.com'])
pattern = r'([A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,4})'

# Trích xuất các chuỗi khớp với pattern
extracted_emails = emails.str.extract(pattern, expand=False).dropna()

print("Dữ liệu email gốc:")
print(emails)

print("\nĐịa chỉ email hợp lệ được trích xuất:")
print(extracted_emails)
```

```python
import pandas as pd

fruit = pd.Series(np.random.choice(['apple', 'banana', 'carrot'], 10))
weights = pd.Series(np.linspace(1, 10, 10))
print(weights.tolist())
print(fruit.tolist())
df = pd.DataFrame({'Fruit': fruit, 'Weight': weights})
average_weight = df.groupby('Fruit')['Weight'].mean()
print("--- Data Frame Tạm thời ---")
print(df)
print("\n--- Khối lượng trung bình của từng loại quả ---")
print(average_weight)
```

```python
import pandas as pd

data = {
    'Team': ['England', 'France', 'Spain', 'Italy', 'Germany', 'Russia', 'Greece', 'Portugal', 'Sweden', 'Poland', 'Croatia', 'Denmark', 'Ireland', 'Ukraine', 'Czech Republic', 'Netherlands'],
    'Goals': np.random.randint(1, 10, 16),
    'Yellow Cards': np.random.randint(0, 5, 16),
    'Red Cards': np.random.randint(0, 2, 16),
    'Shooting Accuracy': [30.1, 45.2, 55.6, 40.5, 38.9, 42.1, 35.0, 48.0, 31.0, 33.3, 39.5, 30.0, 32.5, 36.5, 41.0, 37.5] # Giả định là %
}
euro12 = pd.DataFrame(data)
#1
print("1. Giá trị cột 'Goals':")
print(euro12['Goals'])
#2
num_teams = euro12.shape[0] # Lấy số hàng
print(f"\n2. Số đội tham gia Euro 2012: {num_teams}")
#3
print("\n3. Thông tin DataFrame (euro12.info()):")
euro12.info()
#4
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("\n4. DataFrame 'discipline':")
print(discipline.head())
#5
discipline_sorted = discipline.sort_values(
    by=['Red Cards', 'Yellow Cards'],
    ascending=[False, False] 
)
print("\n5. DataFrame 'discipline' đã sắp xếp:")
print(discipline_sorted.head())
#6a
mean_yellow_cards = euro12['Yellow Cards'].mean()
print(f"\n6. a) Trung bình 'Yellow Cards' của Euro 2012: {mean_yellow_cards:.2f}")
#6b
teams_over_6_goals = euro12[euro12['Goals'] > 6]
print("\n7. b) Các đội đã ghi hơn 6 bàn thắng:")
print(teams_over_6_goals[['Team', 'Goals']])
#7

teams_start_G = euro12[euro12['Team'].str.startswith('G')]
print("\n8. Các đội có tên bắt đầu bằng 'G':")
print(teams_start_G['Team'])
#8

cols_to_keep = euro12.columns[:-3]
euro12_less_3_cols = euro12[cols_to_keep]
print("\n9. DataFrame bỏ 3 cột cuối:")
print(euro12_less_3_cols.head())
#9
selected_cols = ['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']
euro12_selected = euro12[selected_cols]
print("\n10. Các cột được chọn:")
print(euro12_selected.head())
#10
target_teams = ['England', 'Italy', 'Russia']
filtered_rows = euro12[euro12['Team'].isin(target_teams)]
final_output = filtered_rows[['Team', 'Shooting Accuracy']]
print("\n11. 'Team', 'Shooting Accuracy' của England, Italy, Russia:")
print(final_output)
```
