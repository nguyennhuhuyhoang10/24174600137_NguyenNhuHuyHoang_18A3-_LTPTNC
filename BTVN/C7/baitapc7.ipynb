```python
import tkinter as tk
from tkinter import messagebox

def submit():
    name = entry_name.get()
    sid = entry_id.get()
    pwd = entry_pass.get()

    if name and sid and pwd:
        messagebox.showinfo(
            "Thông báo",
            f"Đã gửi thông tin:\nTên: {name}\nID: {sid}"
        )
    else:
        messagebox.showwarning("Lỗi", "Vui lòng nhập đầy đủ thông tin!")

root = tk.Tk()
root.title("Bai 7.4 - Form sinh vien")
root.geometry("350x220")

tk.Label(root, text="Student Name:").grid(row=0, column=0, padx=10, pady=5)
tk.Label(root, text="Student ID:").grid(row=1, column=0, padx=10, pady=5)
tk.Label(root, text="Password:").grid(row=2, column=0, padx=10, pady=5)

entry_name = tk.Entry(root)
entry_id = tk.Entry(root)
entry_pass = tk.Entry(root, show="*")

entry_name.grid(row=0, column=1)
entry_id.grid(row=1, column=1)
entry_pass.grid(row=2, column=1)

btn = tk.Button(root, text="Submit", command=submit)
btn.grid(row=3, column=1, pady=15)

root.mainloop()
```

```python
import tkinter as tk
from PIL import Image, ImageTk
import urllib.request
import io

def load_image():
    path = entry.get()
    try:
        if path.startswith("http"):
            data = urllib.request.urlopen(path).read()
            image = Image.open(io.BytesIO(data))
        else:
            image = Image.open(path)

        image = image.resize((400, 250))
        photo = ImageTk.PhotoImage(image)
        label.config(image=photo)
        label.image = photo
    except:
        label.config(text="Không đọc được ảnh")

root = tk.Tk()
root.title("Chương trình xem ảnh")

entry = tk.Entry(root, width=50)
entry.pack()

tk.Button(root, text="Load Image", command=load_image).pack()

label = tk.Label(root)
label.pack()

root.mainloop()
```

```python
import tkinter as tk

def reverse_string():
    s = entry.get()
    rev = ""
    for i in range(len(s)-1, -1, -1):
        rev += s[i]
    result.config(text=rev)

root = tk.Tk()
root.title("Reverse String")

tk.Label(root, text="Enter a word:").pack()
entry = tk.Entry(root, width=30)
entry.pack()

tk.Button(root, text="Validate", command=reverse_string).pack()
result = tk.Label(root)
result.pack()

root.mainloop()
```

```python
import tkinter as tk

def calc_sum():
    n = int(entry.get())
    s = 0
    for i in range(1, n+1):
        s += i
    result.config(text=f"The sum is: {s}")

root = tk.Tk()
root.title("Sum N")

tk.Label(root, text="Enter value of integer N:").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Validate", command=calc_sum).pack()
result = tk.Label(root)
result.pack()

root.mainloop()
```

```python
import tkinter as tk

def divisors():
    n = int(entry.get())
    d = []
    for i in range(1, n+1):
        if n % i == 0:
            d.append(str(i))
    result.config(text="Divisors: " + " ".join(d))

root = tk.Tk()
root.title("Divisors")

tk.Label(root, text="Enter value of N:").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Validate", command=divisors).pack()
result = tk.Label(root)
result.pack()

root.mainloop()
```

Bài 7.9

```python
import tkinter as tk
from math import gcd

def calculate():
    a = int(e1.get())
    b = int(e2.get())
    if choice.get() == "GCD":
        result.config(text=f"GCD(m,n) = {gcd(a,b)}")
    else:
        lcm = a*b // gcd(a,b)
        result.config(text=f"LCM(m,n) = {lcm}")

root = tk.Tk()
root.title("GCD & LCM")

tk.Label(root, text="Enter value of m:").pack()
e1 = tk.Entry(root)
e1.pack()

tk.Label(root, text="Enter value of n:").pack()
e2 = tk.Entry(root)
e2.pack()

choice = tk.StringVar(value="GCD")
tk.OptionMenu(root, choice, "GCD", "LCM").pack()

tk.Button(root, text="Calculate", command=calculate).pack()
result = tk.Label(root)
result.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def check_age():
    name = name_entry.get()
    age = int(age_entry.get())
    if age >= 18:
        msg = f"Xin chào {name}\nBạn đã trưởng thành!"
    else:
        msg = f"Xin chào {name}\nBạn còn nhỏ tuổi!"
    messagebox.showinfo("Thông báo", msg)

root = tk.Tk()
root.title("User Info")

tk.Label(root, text="Tên:").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Tuổi:").pack()
age_entry = tk.Entry(root)
age_entry.pack()

tk.Button(root, text="OK", command=check_age).pack()
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

con_vat = ["Thân","Dậu","Tuất","Hợi","Tý","Sửu","Dần","Mão","Thìn","Tỵ","Ngọ","Mùi"]

def convert():
    year = int(entry.get())
    animal = con_vat[year % 12]
    messagebox.showinfo("Kết quả", f"Năm {year} là năm {animal}")

root = tk.Tk()
root.title("Zodiac")

tk.Label(root, text="Enter year:").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Convert", command=convert).pack()
root.mainloop()
```

```python
import tkinter as tk

def bmi():
    w = float(weight.get())
    h = float(height.get())
    b = w / (h*h)
    if b < 18.5:
        k = "Gầy"
    elif b < 25:
        k = "Bình thường"
    else:
        k = "Thừa cân"
    result.config(text=f"BMI: {b:.2f}\nKết luận: {k}")

root = tk.Tk()
root.title("BMI Calculator")

tk.Label(root, text="Cân nặng (kg):").pack()
weight = tk.Entry(root)
weight.pack()

tk.Label(root, text="Chiều cao (m):").pack()
height = tk.Entry(root)
height.pack()

tk.Button(root, text="Tính BMI", command=bmi).pack()
result = tk.Label(root)
result.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show():
    messagebox.showinfo("Lập trình hướng đối tượng",
                        "Viết phương trình giải phương trình bậc 2")

root = tk.Tk()
root.title("Hệ thống bài tập Python")

menu = tk.Menu(root)
root.config(menu=menu)

c1 = tk.Menu(menu, tearoff=0)
menu.add_cascade(label="Chương 1", menu=c1)
c1.add_command(label="1.1", command=show)

root.mainloop()
```

```python
import tkinter as tk

def taxi():
    km = float(entry.get())
    money = 12000
    if km > 0.8:
        km -= 0.8
        if km <= 30:
            money += km * 15300
        else:
            money += 30 * 15300 + (km-30) * 12100
    result.config(text=f"Tổng tiền: {int(money)} đồng")

root = tk.Tk()
root.title("Taxi")

tk.Label(root, text="Số km:").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Tính tiền", command=taxi).pack()
result = tk.Label(root)
result.pack()

root.mainloop()
```
