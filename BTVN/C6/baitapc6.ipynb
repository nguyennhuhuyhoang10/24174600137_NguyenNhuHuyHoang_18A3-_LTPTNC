```python
import threading
import time

def worker(thread_id):
    thread_name = threading.current_thread().name
    print(f"Luồng {thread_id}: Tên luồng là {thread_name}")
    time.sleep(0.5) 

def bai_6_1_tao_va_in_ten_luong():
    threads = []
    num_threads = 5
    print(f"--- TẠO VÀ IN TÊN {num_threads} LUỒNG ---")

    for i in range(num_threads):
        thread = threading.Thread(target=worker, args=(i+1,), name=f"Thread-{i+1}")
        threads.append(thread)
        thread.start()
    for thread in threads:
        thread.join()

    print("Tất cả các luồng đã hoàn thành.")
bai_6_1_tao_va_in_ten_luong()
```

```python
import threading
import time
import random

def download_file(file_name):
    thread_name = threading.current_thread().name
    download_time = random.uniform(1, 3) 
    
    print(f"[{thread_name}] Bắt đầu tải xuống: {file_name}. Thời gian dự kiến: {download_time:.2f}s")
    time.sleep(download_time)
    print(f"[{thread_name}]  Hoàn thành tải xuống: {file_name}.")

def bai_6_2_tai_xuong_dong_thoi():
    files_to_download = ["File_A.zip", "Image_B.jpg", "Document_C.pdf", "Video_D.mp4"]
    threads = []
    
    print(f"--- BÀI 6.2: MÔ PHỎNG TẢI XUỐNG ĐỒNG THỜI {len(files_to_download)} TỆP ---")

    start_time = time.time()
    
    for file in files_to_download:
        thread = threading.Thread(target=download_file, args=(file,))
        threads.append(thread)
        thread.start()
        
    for thread in threads:
        thread.join()

    end_time = time.time()
    print(f"\nTổng thời gian hoàn thành: {end_time - start_time:.2f} giây.")
bai_6_2_tai_xuong_dong_thoi()
```

```python
import threading

def tim_so_chan(start, end):
    chan_list = [i for i in range(start, end + 1) if i % 2 == 0]
    print(f"\n[Luồng CHẴN] Các số chẵn từ {start} đến {end}: {chan_list}")

def tim_so_le(start, end):
    le_list = [i for i in range(start, end + 1) if i % 2 != 0]
    print(f"\n[Luồng LẺ] Các số lẻ từ {start} đến {end}: {le_list}")

def bai_6_3_tim_so_chan_le():
    start_range = 30
    end_range = 50
    print(f"--- TÌM SỐ CHẴN/LẺ (TỪ {start_range} ĐẾN {end_range}) BẰNG HAI LUỒNG ---")
    t_chan = threading.Thread(target=tim_so_chan, args=(start_range, end_range), name="Chan_Thread")
    
    t_le = threading.Thread(target=tim_so_le, args=(start_range, end_range), name="Le_Thread")
    
    t_chan.start()
    t_le.start()

    t_chan.join()
    t_le.join()

    print("\n Cả hai luồng đã hoàn thành việc tìm kiếm.")

bai_6_3_tim_so_chan_le()
```

```python
import threading
import requests
import time
URLS = [
    "https://jsonplaceholder.typicode.com/posts/1",
    "https://jsonplaceholder.typicode.com/users/2",
    "https://jsonplaceholder.typicode.com/comments/3",
    "http://httpbin.org/delay/1" 
]

def fetch_url(url):
    thread_name = threading.current_thread().name
    try:
        start_time = time.time()
        response = requests.get(url, timeout=5) 
        end_time = time.time()
        
        status = response.status_code
        length = len(response.text)
        
        print(f"[{thread_name}]  {url}: Status={status}, Kích thước={length} bytes, Thời gian={end_time - start_time:.2f}s")
        
    except requests.exceptions.RequestException as e:
        print(f"[{thread_name}]  Lỗi khi truy cập {url}: {e}")

def bai_6_5_http_dong_thoi():
    threads = []
    print(f"---  THỰC HIỆN {len(URLS)} YÊU CẦU HTTP ĐỒNG THỜI ---")

    start_total = time.time()

    for url in URLS:
        thread = threading.Thread(target=fetch_url, args=(url,))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()

    end_total = time.time()
    print(f"\nTổng thời gian hoàn thành (Đa luồng): {end_total - start_total:.2f} giây.")

bai_6_5_http_dong_thoi()
```

```python
import threading
import time 
lock = threading.Lock()
can_print_even = True 

def in_so_chan(max_num):
    global can_print_even
    for i in range(2, max_num + 1, 2):
        with lock:
            while not can_print_even:
                lock.release() 
                time.sleep(0.001)
                lock.acquire()
                
            print(f"[Luồng CHẴN] {i}")
            can_print_even = False 
def in_so_le(max_num):
    global can_print_even
    for i in range(1, max_num + 1, 2):
        with lock:
            while can_print_even:
                lock.release()
                time.sleep(0.001)
                lock.acquire()
                
            print(f"[Luồng LẺ] {i}")
            can_print_even = True 

def bai_6_6_in_xen_ke_chan_le(max_num=20):
    global can_print_even, lock
    can_print_even = True 
    lock = threading.Lock()
    
    print(f"--- BÀI 6.6: IN SỐ CHẴN/LẺ XEN KẼ ĐẾN {max_num} ---")

    t_chan = threading.Thread(target=in_so_chan, args=(max_num,), name="Even_Thread")
    t_le = threading.Thread(target=in_so_le, args=(max_num,), name="Odd_Thread")
    
    t_le.start() 
    t_chan.start() 
    
    t_chan.join()
    t_le.join()

    print("\n Quá trình in xen kẽ đã hoàn thành.")
bai_6_6_in_xen_ke_chan_le(10)
```

```python
import threading
import time
import random
from datetime import datetime

MAX_RUNTIME = 10 

class WebCrawler(threading.Thread):
    def __init__(self, name, interval):
        super().__init__(name=name)
        self.stop_event = threading.Event()
        self.interval = interval 
        self.start_time = time.time()
        self.print_count = 0

    def run(self):
        print(f"Starting {self.name}")

        while not self.stop_event.is_set():
            if time.time() - self.start_time >= MAX_RUNTIME + random.uniform(0, 3):
                break
            
            current_time_str = datetime.now().strftime("Web %b %d %H:%M:%S %Y")
            print(f"{self.name}: {current_time_str}")
            self.print_count += 1
            
            time.sleep(self.interval + random.uniform(0, 0.5)) 

        print(f"Exiting {self.name}")

def bai_6_7_thuc_thi_nhieu_luong():
    print("Starting Main Thread")
    threads = [
        WebCrawler(name="Google", interval=1.5),  
        WebCrawler(name="Yahoo", interval=2.0),
        WebCrawler(name="Facebook", interval=2.5) 
    ]

    for t in threads:
        t.start()

    for t in threads:
        t.join()

    print("Exiting Main Thread")

bai_6_7_thuc_thi_nhieu_luong()
```

```python
import threading
import random
import math

partial_sums = []
lock = threading.Lock()

def calculate_partial_sum(data_chunk, thread_id):
    current_sum = sum(data_chunk)
    with lock:
        partial_sums.append(current_sum)
    
    data_str = "; ".join(map(str, data_chunk))
    print(f"Thread {thread_id}: {data_str};")


def bai_6_8_tinh_tong_da_luong():
    global partial_sums
    try:
        n = int(input("Nhập số phần tử: ")) 
        num_threads = int(input("Nhập vào số thread: ")) 
    except ValueError:
        print("Lỗi: Đầu vào phải là số nguyên.")
        return

    main_list = [random.randint(0, 10) for _ in range(n)]
    
    print(f"List: {main_list}")
    
    list_length = len(main_list)
    chunk_size = math.ceil(list_length / num_threads) 
    
    threads = []
    partial_sums = [] 
    for i in range(num_threads):
        start_index = i * chunk_size
        end_index = min((i + 1) * chunk_size, list_length)
        
        if start_index >= list_length:
            break
            
        data_chunk = main_list[start_index:end_index]
        
        thread = threading.Thread(
            target=calculate_partial_sum, 
            args=(data_chunk, i + 1), 
            name=f"SumThread_{i + 1}"
        )
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()

    final_total = sum(partial_sums)
    expected_total = sum(main_list)
    print(f"Kiểm tra:{expected_total}")
    
    print(f"\nTổng list={final_total}")
bai_6_8_tinh_tong_da_luong()
```

```python
import threading
import random
import math

partial_maxes = []
lock = threading.Lock()

def calculate_partial_max(data_chunk, thread_id):
    if not data_chunk:
        return

    local_max = max(data_chunk)
    
    with lock:
        partial_maxes.append(local_max)
    
    data_str = "; ".join(map(str, data_chunk))
    print(f"Thread {thread_id}: {data_str}; -> Max cục bộ: {local_max}")


def bai_6_10_tim_max_da_luong():
    global partial_maxes

    try:
        n = int(input("Nhập số phần tử (N): ")) 
        num_threads = int(input("Nhập vào số thread: ")) 
    except ValueError:
        print("Lỗi: Đầu vào phải là số nguyên.")
        return

    main_list = [random.randint(0, 100) for _ in range(n)]
    
    print(f"\nList ({n} phần tử): {main_list}")
    
    list_length = len(main_list)
    chunk_size = math.ceil(list_length / num_threads) 
    
    threads = []
    partial_maxes = [] 

    for i in range(num_threads):
        start_index = i * chunk_size
        end_index = min((i + 1) * chunk_size, list_length)
        
        if start_index >= list_length:
            break
            
        data_chunk = main_list[start_index:end_index]
        
        thread = threading.Thread(
            target=calculate_partial_max, 
            args=(data_chunk, i + 1), 
            name=f"MaxThread_{i + 1}"
        )
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()
    if partial_maxes:
        final_max = max(partial_maxes)
    else:
        final_max = "Không tìm thấy"
    expected_max = max(main_list) if main_list else "Không tìm thấy"
    print(f"Kiểm tra bằng hàm max() chuẩn :{expected_max}")
    
    print(f"\nMax cục bộ tìm được bởi các Threads: {partial_maxes}")
    print(f"Giá trị lớn nhất trong list (Tìm bằng đa luồng) = {final_max}")
bai_6_10_tim_max_da_luong()
```
